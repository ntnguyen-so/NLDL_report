```python
# Import libraries
import numpy as np
import scipy.stats as stats

cutoff = 10000

# Define parameters
w = 3
window_size = 8000/10*w # size of the sliding window for short-term average

alpha = 0.01 # decay factor for EMA
threshold = 0.16 # fixed threshold for Q-function

# Define prediction errors
prediction_errors = adapad_result["err"].values[:cutoff] - adapad_result["threshold"].values[:cutoff]

# Initialize variables
mu = prediction_errors[0] # long-term mean of prediction errors
sigma = prediction_errors[0] # long-term standard deviation of prediction errors
anomalies = [] # list of anomalies

# Loop over the prediction errors
for i in range(len(prediction_errors)):
    # Compute the short-term average of prediction errors using a sliding window
    if i < window_size:
        mu = np.mean(prediction_errors[:i+1])
    else:
        mu = np.mean(prediction_errors[i-window_size+1:i+1])
        
    # calculate mu_tilde
    if i < w:
        mu_tilde = np.mean(prediction_errors[:i+1])
    else:
        mu_tilde = np.mean(prediction_errors[i-w+1:i+1])

    # Compute the Q-function value using the standard normal distribution
    Q_value = stats.norm.sf((mu_tilde - mu) / sigma) 
    
    print(f"{adapad_result.err.values[i]}, {adapad_result.anomalous.values[i]}, {1 - Q_value:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Sample prediction errors and thresholds (example data)
prediction_errors = np.array([0.2, 0.5, 0.8, 1.1, 1.5, 1.9])
thresholds = np.array([0.5, 0.5, 0.8, 0.9, 1.2, 1.4])

# Step 1: Calculate absolute prediction errors
abs_errors = np.abs(prediction_errors)

# Step 2: Normalize prediction errors and thresholds
normalized_errors = (abs_errors - np.min(abs_errors)) / (np.max(abs_errors) - np.min(abs_errors))
normalized_thresholds = (thresholds - np.min(thresholds)) / (np.max(thresholds) - np.min(thresholds))

# Step 3: Assign binary labels based on thresholds
labels = np.where(normalized_errors > normalized_thresholds, 1, 0)

# Step 4: Calculate ROC curve
fpr, tpr, _ = roc_curve(labels, normalized_errors)
roc_auc = auc(fpr, tpr)

# Step 6: Convert distance to classification probabilities
def convert_to_probability(error, thresholds, tpr):
    try:
        idx = np.searchsorted(thresholds, error)
        print(idx)
        if idx == len(thresholds):
            return 1.0
        elif idx == 0:
            return tpr[0]
        else:
            x_interp = np.interp(error, [thresholds[idx-1], thresholds[idx]], [tpr[idx-1], tpr[idx]])
            return x_interp
    except:
        pass

# Step 7: Example usage
probabilities = [convert_to_probability(error, normalized_thresholds, tpr) for error in normalized_errors]
print("Classification Probabilities:", probabilities)
```

```python
def remove_long_shutdown(numbers, num_consecutive, missing_label):
    chunks = []
    current_chunk = []

    i = 0
    while i < len(numbers)-1:
        num = numbers[i]
        if num != missing_label:
            current_chunk.append(i)
        else:
            j = i+1
            while j < len(numbers):
                if numbers[j] == missing_label:
                    j += 1
                else:
                    break

            if j-i < num_consecutive:
                current_chunk += range(i,min(j+1, len(numbers)))# numbers[i:j+1]
            else:
                chunks.append(current_chunk)
                current_chunk = []

            i=j         

        i+= 1

    # Append the last chunk
    if current_chunk:
        chunks.append(current_chunk)

    to_ret = []
    # Print the chunks
    for i, chunk in enumerate(chunks, 1):
        to_ret += chunk
        
    return to_ret
```
