```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
num_col = 10
df = pd.read_csv('./_immediate/Tide_pressure.csv')
#df = np.where(df['is_anomaly'] == 1, np.nan, df['value'])
df = np.where(df['is_anomaly'] == 1, np.nan, df['value'])
df = df.reshape((-1, 1))
df = np.tile(df, (1, num_col))

df = StandardScaler().fit_transform(df)
output_file = './oceantidepressure/oceantidepressure_normal.txt'
np.savetxt(output_file, df, delimiter=' ', newline='\n', fmt='%1.3f')

# Open the file in read mode and read its content
with open(output_file, 'r') as file:
    file_content = file.read()

# Replace 'nan' with 'NaN' in the content
modified_content = file_content.replace('nan', 'NaN')

# Open the same file in write mode and write the modified content back
with open(output_file, 'w') as file:
    file.write(modified_content)

#df['value'].to_frame()
```

```python
num_col = 10
df = pd.read_csv('./_immediate/Tide_pressure.csv')
#df = np.where(df['is_anomaly'] == 1, np.nan, df['value'])
df = np.where(df['is_anomaly'] == 1, np.nan, df['value'])
df = df.reshape((-1, 1))
df = np.tile(df, (1, num_col))
for i in range(1, num_col):
    df[:-i, i] = df[i:, i]
    df = df[i:]

df = df[num_col:]
print(df)
df = StandardScaler().fit_transform(df)
#df
output_file = './oceantidepressuremissingshift/oceantidepressuremissingshift_normal.txt'
np.savetxt(output_file, df, delimiter=' ', newline='\n', fmt='%1.3f')

# Open the file in read mode and read its content
with open(output_file, 'r') as file:
    file_content = file.read()

# Replace 'nan' with 'NaN' in the content
modified_content = file_content.replace('nan', 'NaN')

# Open the same file in write mode and write the modified content back
with open(output_file, 'w') as file:
    file.write(modified_content)

#df['value'].to_frame()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Create figure and plot lines
fig = go.Figure()

# Add line plot for 'value'
fig.add_trace(go.Scatter(x=df['timestamp'], y=df['value'], mode='lines', name='Value'))
#fig.add_trace(go.Scatter(x=np.arange(len(df)), y=df['value'], mode='lines', name='Value'))

# Add scatter plot for 'is_anomaly'
anomaly_points = df[df['is_anomaly'] == 1]
fig.add_trace(go.Scatter(x=anomaly_points['timestamp'], y=anomaly_points['value'], mode='markers', name='Anomaly', marker=dict(color='red', size=10)))
#fig.add_trace(go.Scatter(x=np.arange(len(df)), y=anomaly_points['value'], mode='markers', name='Anomaly', marker=dict(color='red', size=10)))

# Update layout
fig.update_layout(title='Value Over Time with Anomalies',
                  xaxis_title='Timestamp',
                  yaxis_title='Value')

# Show plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

start_range, end_range = 7960, 8001
df['timestamp_'] = pd.to_datetime(df.iloc[start_range:end_range, 0])  # Convert timestamp column to datetime
df['value_'] = df.iloc[start_range:end_range, 1]
df['is_anomaly_'] = df.iloc[start_range:end_range, 2]

# Plotting
plt.figure(figsize=(10, 6))

# Plot value over time
plt.plot(df['timestamp_'], df['value_'], color='blue', label='Value')

# Highlight anomalies
anomaly_points = df[df['is_anomaly_'] == 1]
plt.scatter(anomaly_points['timestamp_'], anomaly_points['value_'], color='red', label='Anomaly')

plt.title('Tide Pressure data with Anomalies')
plt.xlabel('Timestamp')
plt.ylabel('Value (kPa)')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show plot
plt.show()
```
